```python
import pandas as pd
import numpy as np



# Replace placeholder codes
df = df.replace("-1", np.nan)
df = df.replace("-3", np.nan)
df = df.replace(-1, np.nan)
df = df.replace(-3, np.nan)

# ----------------------------
# 1. Clean city
# ----------------------------
df["city_clean"] = df["8 Identify City "].astype("string").str.strip()

# ----------------------------
# 2. Clean neighbourhood
# ----------------------------
df["neighbourhood_clean"] = pd.Series(pd.NA, index=df.index, dtype="string")

df.loc[df["city_clean"] == "Johannesburg", "neighbourhood_clean"] = (
    df.loc[df["city_clean"] == "Johannesburg", "9 Johannesburg Neighbourhoods"]
    .astype("string")
    .str.strip()
)

df.loc[df["city_clean"] == "Accra", "neighbourhood_clean"] = (
    df.loc[df["city_clean"] == "Accra", "9 Accra Neighbourhoods "]
    .astype("string")
    .str.strip()
)

df.loc[df["city_clean"] == "Nairobi", "neighbourhood_clean"] = (
    df.loc[df["city_clean"] == "Nairobi", "9 Nairobi Neighbourhoods"]
    .astype("string")
    .str.strip()
)

# Remove accidental string placeholders if any remain
df["neighbourhood_clean"] = df["neighbourhood_clean"].replace({
    "nan": pd.NA,
    "-1": pd.NA,
    "-3": pd.NA
})

# ----------------------------
# 3. Clean migrant status
# ----------------------------
df["migrant_status_clean"] = df["3 Migrant"].map({
    "Yes": "Born outside country",
    "No": "Born in country"
})

# ----------------------------
# 4. Create age group
# ----------------------------
current_year = 2021
df["age"] = current_year - pd.to_numeric(df["104 Year of Birth"], errors="coerce")

def age_group_func(age):
    if pd.isna(age):
        return pd.NA
    elif age < 25:
        return "18–24"
    elif age < 35:
        return "25–34"
    elif age < 45:
        return "35–44"
    elif age < 55:
        return "45–54"
    else:
        return "55+"

df["age_group"] = df["age"].apply(age_group_func)

# ----------------------------
# Inspect results
# ----------------------------
print("\n=== CITY CLEAN ===")
print(df["city_clean"].value_counts(dropna=False))

print("\n=== NEIGHBOURHOOD CLEAN ===")
print(df["neighbourhood_clean"].value_counts(dropna=False))

print("\n=== MIGRANT STATUS CLEAN ===")
print(df["migrant_status_clean"].value_counts(dropna=False))

print("\n=== AGE GROUP ===")
print(df["age_group"].value_counts(dropna=False))
```

```python
# ----------------------------
# Clean country of origin
# ----------------------------

df["country_of_origin_clean"] = (
    df["103 Country of Origin"]
    .astype("string")
    .str.strip()
    .str.title()
)

# Fix known variants manually if needed
df["country_of_origin_clean"] = df["country_of_origin_clean"].replace({
    "Democratic Republic Of The Congo": "Democratic Republic of the Congo",
    "Drc": "Democratic Republic of the Congo",
    "Ivory Coast": "Cote D'Ivoire",
    "Côte D'Ivoire": "Cote D'Ivoire"
})

print("\n=== COUNTRY OF ORIGIN CLEAN ===")
print(df["country_of_origin_clean"].value_counts().head(20))

print("\n=== UNIQUE COUNTRIES COUNT ===")
print(df["country_of_origin_clean"].nunique())
```

```python
# ---------------------------------
# Collapse small country categories
# ---------------------------------

country_counts = df["country_of_origin_clean"].value_counts()

# Identify countries with fewer than 10 respondents
small_countries = country_counts[country_counts < 10].index

df["country_of_origin_grouped"] = df["country_of_origin_clean"].replace(
    small_countries, "Other"
)

print("\n=== GROUPED COUNTRY COUNTS ===")
print(df["country_of_origin_grouped"].value_counts())

print("\n=== UNIQUE GROUPED COUNTRIES ===")
print(df["country_of_origin_grouped"].nunique())
```

```python
# ----------------------------
# Household size grouping
# ----------------------------

df["household_members"] = pd.to_numeric(df["133 Household Members"], errors="coerce")

def hh_group(x):
    if pd.isna(x):
        return pd.NA
    elif x == 1:
        return "1"
    elif x in [2, 3]:
        return "2–3"
    elif x in [4, 5]:
        return "4–5"
    else:
        return "6+"

df["household_size_group"] = df["household_members"].apply(hh_group)

print("\n=== HOUSEHOLD SIZE GROUP ===")
print(df["household_size_group"].value_counts(dropna=False))
```

```python
# ----------------------------
# Standardise trust variables
# ----------------------------

trust_vars = [
    "421 Local Police",
    "422 Local Councillor",
    "423 Mayor",
    "424 Chief or Traditional Leaders",
    "425 Local",
    "426 Local Country People",
    "427 Neighbours",
    "428 Local Service Providers",
    "429 Religious Leaders",
    "430 Other religious leaders",
    "431 Business people",
    "432 Judges"
]

def clean_trust(x):
    if pd.isna(x):
        return pd.NA
    
    x = str(x).strip()
    
    if x in ["Yes"]:
        return "Yes"
    elif x in ["No"]:
        return "No"
    elif x in ["Yes & No", "Yes and No"]:
        return "Mixed"
    elif x in ["DK", "Don't know"]:
        return "Don't know"
    else:
        return pd.NA

for col in trust_vars:
    if col in df.columns:
        df[col + "_clean"] = df[col].apply(clean_trust)

# Inspect one example
print("\n=== CLEANED TRUST POLICE ===")
print(df["421 Local Police_clean"].value_counts(dropna=False))
```

```python
# ----------------------------
# Clean discrimination variables
# ----------------------------

discrim_vars = [
    "543 Discrimination based on sex",
    "544 Discrimination based on religion",
    "545 Discrimination based on ethnicity ",
    "546 Discrimination based on where you come from",
    "547 Discrimination based on economic status",
    "548 Political Discrimination"
]

def clean_discrimination(x):
    if pd.isna(x):
        return pd.NA
    
    x = str(x).strip()
    
    mapping = {
        "Never": "Never",
        "Once or twice": "Once or twice",
        "Several Times": "Several times",
        "Several times": "Several times",
        "Many times": "Many times",
        "Not applicable": "Not applicable",
        "DK/RA": "Don't know",
        " DK/RA": "Don't know"
    }
    
    return mapping.get(x, pd.NA)

for col in discrim_vars:
    if col in df.columns:
        df[col + "_clean"] = df[col].apply(clean_discrimination)

# Inspect one
print("\n=== CLEANED DISCRIMINATION (Ethnicity example) ===")
print(df["545 Discrimination based on ethnicity _clean"].value_counts(dropna=False))
```

```python
import pandas as pd

summary_vars = [
    "belong_here",
    "perceived_outsider",
    "trust_police",
    "trust_councillor",
    "trust_neighbours",
    "stay_or_go",
    "medium_term_goal",
    "long_term_goal"
]

summary_rows = []

for var in summary_vars:
    counts = public_df[var].value_counts(dropna=False)
    total = counts.sum()
    
    for category, n in counts.items():
        summary_rows.append({
            "geography_level": "overall",
            "city": "All",
            "neighbourhood": "All",
            "variable": var,
            "category": category if pd.notna(category) else "Missing",
            "n": n,
            "pct": round((n / total) * 100, 2)
        })

summary_overall = pd.DataFrame(summary_rows)

print("=== OVERALL SUMMARY SAMPLE ===")
print(summary_overall.head(20))
```

```python
summary_rows = []

for var in summary_vars:
    for city in public_df["city"].unique():
        df_city = public_df[public_df["city"] == city]
        counts = df_city[var].value_counts(dropna=False)
        total = counts.sum()
        
        for category, n in counts.items():
            summary_rows.append({
                "geography_level": "city",
                "city": city,
                "neighbourhood": "All",
                "variable": var,
                "category": category if pd.notna(category) else "Missing",
                "n": n,
                "pct": round((n / total) * 100, 2)
            })

summary_city = pd.DataFrame(summary_rows)

print("=== CITY SUMMARY SAMPLE ===")
print(summary_city.head(20))
```

```python
summary_rows = []

for var in summary_vars:
    for city in public_df["city"].unique():
        df_city = public_df[public_df["city"] == city]
        
        for neighbourhood in df_city["neighbourhood"].unique():
            df_nb = df_city[df_city["neighbourhood"] == neighbourhood]
            
            counts = df_nb[var].value_counts(dropna=False)
            total = counts.sum()
            
            for category, n in counts.items():
                summary_rows.append({
                    "geography_level": "neighbourhood",
                    "city": city,
                    "neighbourhood": neighbourhood,
                    "variable": var,
                    "category": category if pd.notna(category) else "Missing",
                    "n": n,
                    "pct": round((n / total) * 100, 2)
                })

summary_neighbourhood = pd.DataFrame(summary_rows)

print("=== NEIGHBOURHOOD SUMMARY SAMPLE ===")
print(summary_neighbourhood.head(20))
```

```python
summary_all = pd.concat(
    [summary_overall, summary_city, summary_neighbourhood],
    ignore_index=True
)

print("=== FULL SUMMARY SHAPE ===")
print(summary_all.shape)

summary_all.head(10)
```
