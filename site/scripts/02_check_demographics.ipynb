```python
# --- Step 2: Create age and age groups ---

# Ensure year of birth is numeric
df["104 Year of Birth"] = pd.to_numeric(df["104 Year of Birth"], errors="coerce")

# Calculate age at survey year
df["age"] = 2021 - df["104 Year of Birth"]

# Create age groups
def age_bucket(x):
    if pd.isna(x):
        return "Missing"
    elif x < 18:
        return "Missing"
    elif x <= 24:
        return "18-24"
    elif x <= 34:
        return "25-34"
    elif x <= 44:
        return "35-44"
    elif x <= 54:
        return "45-54"
    else:
        return "55+"

df["age_group"] = df["age"].apply(age_bucket)

print("\n=== AGE GROUP DISTRIBUTION ===")
print(df["age_group"].value_counts(dropna=False))
```

```python
# --- Step 3: Create clean demographic variables ---

# -------------------------
# Gender
# -------------------------
df["gender"] = df["4 Gender"].astype(str).str.strip()

df["gender"] = df["gender"].replace({
    "Male": "Male",
    "Female": "Female"
})

df.loc[~df["gender"].isin(["Male", "Female"]), "gender"] = "Other / Missing"


# -------------------------
# Migrant status
# -------------------------
df["migrant_status"] = df["3 Migrant"].astype(str).str.strip()

df["migrant_status"] = df["migrant_status"].replace({
    "Yes": "Migrant",
    "No": "Non-migrant"
})

df.loc[~df["migrant_status"].isin(["Migrant", "Non-migrant"]), "migrant_status"] = "Missing"


# -------------------------
# Education grouping
# -------------------------
edu = df["111 Education "].astype(str).str.strip()

df["education_group"] = "Missing"

df.loc[edu.str.contains("No", case=False, na=False), "education_group"] = "No formal schooling"

df.loc[edu.str.contains("Primary", case=False, na=False), "education_group"] = "Primary"

df.loc[edu.str.contains("Secondary", case=False, na=False), "education_group"] = "Secondary"

df.loc[edu.str.contains("University|College|Tertiary", case=False, na=False), "education_group"] = "Tertiary"


# -------------------------
# Employment grouping
# -------------------------
emp = df["303 Current Employment Status"].astype(str).str.strip()

df["employment_status"] = "Missing"

df.loc[emp.str.contains("Self", case=False, na=False), "employment_status"] = "Self-employed"

df.loc[emp.str.contains("Employed", case=False, na=False), "employment_status"] = "Employed"

df.loc[emp.str.contains("Unemployed", case=False, na=False), "employment_status"] = "Unemployed"

df.loc[emp.str.contains("Student", case=False, na=False), "employment_status"] = "Student"


# -------------------------
# Inspect distributions
# -------------------------
print("\n=== GENDER ===")
print(df["gender"].value_counts())

print("\n=== MIGRANT STATUS ===")
print(df["migrant_status"].value_counts())

print("\n=== EDUCATION GROUP ===")
print(df["education_group"].value_counts())

print("\n=== EMPLOYMENT STATUS ===")
print(df["employment_status"].value_counts())
```

```python
# --- Step 5: Create unified city and neighbourhood fields for public summary ---

public_df["city"] = public_df["city"].astype(str).str.strip()

public_df["neighbourhood"] = "Missing"

public_df.loc[public_df["city"] == "Johannesburg", "neighbourhood"] = public_df["jhb_neighbourhood"]
public_df.loc[public_df["city"] == "Accra", "neighbourhood"] = public_df["accra_neighbourhood"]
public_df.loc[public_df["city"] == "Nairobi", "neighbourhood"] = public_df["nairobi_neighbourhood"]

# clean neighbourhood labels
public_df["neighbourhood"] = (
    public_df["neighbourhood"]
    .astype(str)
    .str.replace("–", "-", regex=False)
    .str.replace("\u200b", "", regex=False)
    .str.strip()
)

# replace obvious placeholders
public_df.loc[public_df["neighbourhood"].isin(["-1", "", "nan", "None"]), "neighbourhood"] = "Missing"

print(public_df[["city", "neighbourhood"]].head(10))
print("\nNeighbourhood counts:")
print(public_df["neighbourhood"].value_counts(dropna=False))
```

```python
# --- Step 6: Build survey_public_summary_v2.csv ---

demo_vars = [
    "age_group",
    "gender",
    "migrant_status",
    "education_group",
    "employment_status"
]

summary_parts = []

# neighbourhood-level summaries
for var in demo_vars:
    temp = public_df[[ "city", "neighbourhood", var ]].copy()
    temp = temp.rename(columns={var: "category"})
    
    # standardize category
    temp["category"] = temp["category"].astype(str).str.strip()
    temp.loc[temp["category"].isin(["", "nan", "None"]), "category"] = "Missing"

    counts = (
        temp.groupby(["city", "neighbourhood", "category"], dropna=False)
        .size()
        .reset_index(name="n")
    )

    # small-cell rule
    counts.loc[counts["n"] < 5, "category"] = "Missing"

    # recompute after collapsing
    counts = (
        counts.groupby(["city", "neighbourhood", "category"], dropna=False)["n"]
        .sum()
        .reset_index()
    )

    denom = (
        counts.groupby(["city", "neighbourhood"])["n"]
        .sum()
        .reset_index(name="denominator")
    )

    counts = counts.merge(denom, on=["city", "neighbourhood"], how="left")
    counts["share"] = counts["n"] / counts["denominator"]
    counts["geography_level"] = "neighbourhood"
    counts["variable"] = var

    summary_parts.append(counts)

# city-level summaries
for var in demo_vars:
    temp = public_df[[ "city", var ]].copy()
    temp = temp.rename(columns={var: "category"})
    
    temp["category"] = temp["category"].astype(str).str.strip()
    temp.loc[temp["category"].isin(["", "nan", "None"]), "category"] = "Missing"

    counts = (
        temp.groupby(["city", "category"], dropna=False)
        .size()
        .reset_index(name="n")
    )

    counts.loc[counts["n"] < 5, "category"] = "Missing"

    counts = (
        counts.groupby(["city", "category"], dropna=False)["n"]
        .sum()
        .reset_index()
    )

    denom = (
        counts.groupby(["city"])["n"]
        .sum()
        .reset_index(name="denominator")
    )

    counts = counts.merge(denom, on="city", how="left")
    counts["share"] = counts["n"] / counts["denominator"]
    counts["geography_level"] = "city"
    counts["neighbourhood"] = "All"
    counts["variable"] = var

    summary_parts.append(counts)

summary_v2 = pd.concat(summary_parts, ignore_index=True)

summary_v2 = summary_v2[
    ["geography_level", "city", "neighbourhood", "variable", "category", "n", "denominator", "share"]
].copy()

summary_v2.to_csv(
    r"C:\Users\a0053672\Desktop\urban-mobility-explorer\site\data\survey_public_summary_v2.csv",
    index=False,
    encoding="utf-8-sig"
)

print(summary_v2.head(20))
print("\nRows:", len(summary_v2))
print("\nVariables:")
print(summary_v2["variable"].value_counts())
```
